```python
def pp(df_name):
    ATT = pd.read_csv(df_name)

    # 데이터 process
    ## 필요 없는 컬럼 삭제
    ATT_nunique = ATT.nunique().reset_index()
    drop_cols = []
    # nunique 값이 1개
    cond1 = ATT_nunique[0] <=1
    cols = ATT_nunique.loc[cond1]['index'].tolist()
    print(cols)
    drop_cols = cols
    for col in ATT.columns:
        if 'id' in col:
            drop_cols.append(col)
    
    # isnull 값 139개 이상
    ATT_isnull = ATT.isnull().sum().reset_index()
    cond2 = ATT_isnull[0] >= 139
    cols = ATT_isnull.loc[cond2]['index'].tolist()
    print(cols)
    for col in cols:
        if col not in drop_cols:
            drop_cols.append(col)
    print(drop_cols)
            
            
    ATT_processing = ATT.drop(columns=drop_cols)
    
    ATT_processing = ATT_processing.drop(columns=['intmonth','intday','inthour','intmin'])
    

    
    # 데이터 타입 설정
    ATT_processing = ATT_processing.astype('object')
    # test 할 떄만 ai_accuracy 변환
    ATT_processing['ai_accuracy'] = ATT_processing['ai_accuracy'].astype('int')
    ATT_processing['fltbpsin'] = ATT_processing['fltbpsin'].astype('int')
    ATT_processing['fltbpsout'] = ATT_processing['fltbpsout'].astype('int')
    ATT_processing['fltppsin'] = ATT_processing['fltppsin'].astype('int')
    ATT_processing['fltppsout'] = ATT_processing['fltppsout'].astype('int')
    ATT_processing['strifspeed'] = ATT_processing['strifspeed'].astype('int')

    
    # test 할 때만 ai_accuracy drop
#     ATT_processing = ATT_processing.drop(columns='ai_accuracy')
    
    
    return ATT_processing
```
